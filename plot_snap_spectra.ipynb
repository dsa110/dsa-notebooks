```python
import numpy as np, matplotlib.pyplot as plt
import os
```

```python
os.system('rm -rf spectra')
os.system('cp ../spectra.tar .; tar -xvf spectra.tar')
```

```python
x = 1530.-np.arange(1024)*250./1024.
for snap in ['snap00','snap01','snap02','snap03','snap04','snap05','snap06','snap07']:
    
    plt.figure(figsize=(10,5.5))
    plt.title(snap,fontsize=14)
    plt.xlabel('Frequency (MHz)',fontsize=14)
    plt.ylabel('Power (relative dB)',fontsize=14)
    
    d = np.load('spectra/'+snap+'.npz')
    plt.plot(x,10.*np.log10(d['s1'].real),label=str(d['ants'][0])+'B')
    plt.plot(x,10.*np.log10(d['s2'].real),label=str(d['ants'][0])+'A')
    plt.plot(x,10.*np.log10(d['s3'].real),label=str(d['ants'][1])+'B')
    plt.plot(x,10.*np.log10(d['s4'].real),label=str(d['ants'][1])+'A')
    plt.plot(x,10.*np.log10(d['s5'].real),label=str(d['ants'][2])+'B')
    plt.plot(x,10.*np.log10(d['s6'].real),label=str(d['ants'][2])+'A')
    
    plt.legend(fontsize=14)
    plt.show()
```
